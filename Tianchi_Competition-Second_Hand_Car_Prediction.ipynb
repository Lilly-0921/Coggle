# Step 3: Feature Engineering/特征与标签构建

```python
numerical_cols=Train_data.select_dtypes(exclude='object').columns
numerical_cols
```

```python
categorical_cols = Train_data.select_dtypes(include='object').columns
categorical_cols
```

```python
# select the features 


feature_cols = [col for col in numerical_cols if col not in ['SaleID','name','regDate','creatDate','price','model','brand','regionCode','seller']]
feature_cols = [col for col in feature_cols if 'Type' not in col]

feature_cols
```

1. the name, and whatever date-related info is not useful in our model
2. price is the y-value we are predicting
3. the rest are some first tryout definitely you can change later(add on or remove, etc.)
   Well here the 'bodyType', 'fuelType' and 'offerType' are categorical values which can cause misunderstandings on the model. 

```python
# confirm with the x_data and y_data
X_data = Train_data[feature_cols]
Y_data = Train_data['price']

print('X train shape:', X_data.shape)
print('Y train shape:', Y_data.shape)
```

Train dataset has 31 columns while test data has 30 columns, the difference is the y predictor "price". 

```python
set(Train_data.columns)-set(Test_data.columns)
```

Now tidy up with the "Test" dataset.

Sef define a statistical function

```python
def Sta_inf(data):
    print('_min',np.min(data))
    print('_max:',np.max(data))
    print('_mean',np.mean(data))
    print('_ptp',np.ptp(data))
    print('_std',np.std(data))
    print('_var',np.var(data))
```

```python
print('Statistics of features selected:')
Sta_inf(Y_data)
```
